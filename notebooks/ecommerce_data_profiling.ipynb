# E-commerce Orders — Data Profiling
Quick overview of the raw ecommerce orders dataset.

```python
import pandas as pd
from pathlib import Path

RAW = Path('../data/raw/daily_ecommerce_orders.csv')
df = pd.read_csv(RAW, parse_dates=['order_date'])
print(f'Shape: {df.shape}')
df.head()
```

```python
df.info()
print('\nNull counts:')
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df['order_status'].value_counts(normalize=True).mul(100).round(1).to_frame('pct')
```

```python
df.groupby('product_category')['order_value'].agg(['sum', 'mean', 'count']).sort_values('sum', ascending=False)
```
